##2. SQLite Database Operations:

###2.1

```python
conn = sqlite3.connect('health.db')
c = conn.cursor()
```

I am connecting to sqlite3 and creating a sql database within the notebook.

###2.2 Manual table creation:

```python
c.execute("""
            CREATE TABLE catholichealth
                (
                    hospital_name text,
                    insurance_type text,
                    patient_first_name text,
                    patient_last_name text,
                    blood_pressure_systolic real,
                    blood_pressure_diastolic real
                );
          """)

conn.commit()
```

I am creating a new table within the health.db folder called catholichealth and putting in some column names for patients coming in and getting their blood pressure taken.

```python

c.execute('''
  SELECT name
  FROM sqlite_master
  WHERE type='table';
  ''')

c.fetchall()
```

The table was detected within the database

```python
sql_query = """

INSERT INTO catholichealth (
    'hospital_name',
    'insurance_type',
    'patient_first_name',
    'patient_last_name',
    'blood_pressure_systolic',
    'blood_pressure_diastolic'
  )
  values (
    'good samaritan hospital',
    'fidelis',
    'john',
    'doe',
    120,
    80
  );

"""

print(sql_query)
```

Here I am using SQL to insert values into the columns I created in the table. Filling out both categorical and numerical information.

```python

c.execute(sql_query)
conn.commit()
```

I am then committing these changes into the table.

```python
sql_query_2 = """

select *
from catholichealth;

"""

c.execute(sql_query_2)
print(c.fetchall())
```

Now I am checking to see if the changes were pushed to the table by making the code print out any values in rows.
